```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame(np.random.rand(7, 7), 'a,b,c,d,e,f,g'.split(','), 'S,U,V,W,X,Y,Z'.split(','))
df
```

# Valymas

```python
purvini = df[df>0.15]
purvini
```

```python
purvini.isna()
```

```python
purvini.isnull()
```

```python
purvini.isnull().sum()
```

```python
purvini['X'].isnull().sum()
```

```python
# išmetam visas eilutes, kuriose yra sugadintų (NaN) duomanų
purvini.dropna()
```

```python
# jeigu yra bent 5 sveiki duomenys eilutėje, nemetam
purvini.dropna(thresh=5)
```

```python
# jeigu yra bent 5 sveiki duomenys stulpelyje, nemetam
purvini.dropna(thresh=6, axis=1)
```

```python
# pakeičiame NaN žodžiu kiauras
purvini.fillna('kiauras')
```

```python
print(purvini['S'])
# pakeičiam neegzistuojantį duomenį stulpelio vidurkiu, tik stulpelyje S
purvini['S'].fillna(value=purvini['S'].mean())
```

```python
# užpildom visas NaN reikšmes jų stulpelių vidurkiais
purvini.fillna(purvini.mean())
```

# Grupavimas

```python
duomenys = {
    'šalis': str('Lietuva,'*3+'Latvija,'*3+'Estija,'*3)[:-1].split(','),
    'miestas': 'Vilnius,Kaunas,Klaipėda,Riga,Venspils,Daugpils,Tallinn,Tartu,Pärnu'.split(','),
    'gyventojai': [541, 287, 147, 716, 43, 105, 400, 101, 46]
}
miestai = pd.DataFrame(duomenys)
miestai
```

```python
countries = miestai.groupby('šalis')
countries.mean(numeric_only=True)
```

```python
# išspausdina sugrupuotą žodyną
countries.groups
```

```python
# išspausdina raktažodžių sąrašą
countries.groups.values()
```

```python
countries.sum(numeric_only=True)
```

```python
countries.count()
```

```python
countries.min()
```

```python
countries.max()
```

```python
# konkrečios grupės duomenų traukimas
countries.sum(numeric_only=True).loc['Lietuva']
```

```python
countries.describe()
```

```python
countries.describe().transpose()
```

```python
countries.describe().transpose()['Estija']
```

## DF jungimas

```python
mtop = miestai[:6]
mtop
```

```python
mbot = miestai[6:]
mbot
```

```python
pd.concat([mtop, mbot])
```

```python
pd.concat([mtop, mbot], axis=1)
```

```python
mright = mbot.reset_index()
mright.drop(columns='index', inplace=True)
mright
```

```python
pd.concat([mtop, mright], axis=1)
```

```python
lenkija = {
    "šalis": ",".join(["Lenkija" for x in range(3)]).split(','),
    "miestas": "Varšuva,Vroclavas,Gdanskas".split(','),
    "gyventojai": [1688, 638, 461]
}
polonia = pd.DataFrame(lenkija)
polonia
```

```python
mpl = pd.concat([mright, polonia]).reset_index().drop(columns='index')
mpl
```

```python
sujungta = pd.concat([mtop, mpl], axis=1)
sujungta['šalis']
```

```python
key = 'A,B,C,D,E,F'.split(',')
mtop['key'] = key
mpl['key'] = key
```

```python
pd.merge(mtop, mpl, on='key')
```

```python
key2 = 'B,C,D,E,F,G'.split(',')
mtop['key2'] = key2
mpl['key2'] = key2
```

```python
mtop, mpl
```

```python
pd.merge(mtop, mpl, on=['key', 'key2'])
```

```python
keyx = 'W,C,D,E,F,W'.split(',')
mpl['key2'] = keyx
mtop, mpl
```

```python
mtop.join(mpl, lsuffix='_1', rsuffix='_2').join(mpl)
```
