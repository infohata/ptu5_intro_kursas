```python
import pandas as pd
```

```python
vardai = ['Justina', 'Darius', 'Kęstutis', 'Linas', 'Arnoldas', 'Marius', 'Ingrida']
```

```python
darbuotojai = pd.DataFrame(data=vardai, columns=['Vardas'])
darbuotojai
```

```python
darbuotojai['Atlyginimas'] = [5000, 7000, 3500, 9999, 2000, 3700, 4000]
darbuotojai
```

```python
darbuotojai['Valandos'] = [120, 180, 80, 150, 160, 160, 140]
darbuotojai
```

```python
darbuotojai['Valandinis'] = darbuotojai['Atlyginimas'] / darbuotojai['Valandos']
darbuotojai
```

```python
darbuotojai['Pareigos'] = ['Ekspertė', 'Direktorius', 'Programuotojas', 'Youtuberis', 'Vairuotojas', 'Pardavėjas', 'PR']
darbuotojai
```

```python
darbuotojai['Lytis'] = ['M', 'V', 'V', 'V', 'V', 'V', 'M']
darbuotojai
```

# filtruojam

```python
darbuotojai[(darbuotojai['Atlyginimas'] > 4000) & ((darbuotojai['Valandinis'] > 30) | (darbuotojai['Valandos'] <= 160))]
```

```python
darbuotojai[((darbuotojai['Atlyginimas'] > 4000) & (darbuotojai['Valandinis'] > 30)) | (darbuotojai['Valandos'] <= 160)]
```

```python
darbuotojai[(darbuotojai['Lytis'] == 'V') & ((darbuotojai['Valandinis'] < 40) | (darbuotojai['Atlyginimas'] > 3000))]
```

```python
darbuotojai[((darbuotojai['Lytis'] == 'V') & (darbuotojai['Valandinis'] < 40)) | (darbuotojai['Atlyginimas'] > 3000)]
```
