```python
import pandas as pd
import seaborn as sns

# parsisiunčiam duomenis iš seaborn
mpg = sns.load_dataset('mpg')
mpg.head()
```

## Packing/unpacking Data with Pickle

```python
# kokiu dar duomenu turi Seaborn
sns.get_dataset_names()
```

```python
import pickle
with open('data/mpg.pkl', 'wb') as failas:
    pickle.dump(mpg, failas)
```

```python
with open('data/mpg.pkl', 'rb') as failas:
    mpg = pickle.load(failas)
mpg.tail()
```

## Cleaning Data

```python
mpg.info()
```

```python
mpg[mpg['horsepower'].isnull()]
```

```python
mpg['horsepower'].loc[32] = 75
mpg['horsepower'].loc[126] = 105
mpg['horsepower'].loc[330] = 51
mpg['horsepower'].loc[336] = 120
mpg['horsepower'].loc[354] = 73
mpg['horsepower'].loc[374] = 82
```

```python
dummies = pd.get_dummies(mpg['origin'])
dummies.head()
```

```python
data = pd.concat([mpg, dummies], axis=1)
data.tail()
```

```python
data.drop(['origin', 'name'], axis=1, inplace=True)
data.tail()
```

## Preparing Data

```python
features = data[list(data.columns)[1:]]
features
```

```python
target = data['mpg']
target
```

```python
from sklearn.model_selection import train_test_split
features_train, features_test, target_train, target_test = train_test_split(
    features, target, test_size=0.33, random_state=42
)
features_train
```

```python
from sklearn.linear_model import LinearRegression
```

```python
ai = LinearRegression().fit(features_train, target_train)
```

```python
predictions = ai.predict(features_test)
predictions
```

```python
data_predictions = pd.Series(data=predictions, name="mpg_predictions")
data_predictions
```

```python
comparison = pd.concat([target_test.reset_index(), data_predictions], axis=1)
comparison
```

```python
ai.score(features_test, target_test)
```

```python
final_data = pd.merge(mpg[['name']].reset_index(), comparison, on='index')
final_data
```
